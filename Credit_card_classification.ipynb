Classification of Credit card fraud dataset

Algorithms used :- KNN RandomForest

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler, normalize




path="/content/sample_data/creditcard_2023.csv"
df=pd.read_csv(path)

df.info()
```

Check for null values

```python
df.isna().sum()
df.isnull().sum()
df.dropna()
```

Interpolate null values with forward fill

```python
df=df.fillna(method="ffill")
```

Check for na values ,drop if any

```python
df.isna().sum()
df.isnull().sum()
df.dropna()
```

Split data set into features and class
X -> Features
Y -> Class


```python
x=df.iloc[:,0:29].values
y=df.iloc[:,-1:].values
#labels=df.columns()
#print(x.columns)
#print(">>>>>")
#print(y.columns)
```

Split data set into training and testing data set in the proportion 70 : 30 respectively

```python
#splitting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)

print(len(x_train),len(y_train),len(x_test),len(y_test))

#Feature Scaling
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

Create the model and predict values for test data set

```python
classifier = KNeighborsClassifier(n_neighbors=9)
classifier.fit(x_train, y_train.ravel())
# Predict y data with classifier
y_predict = classifier.predict(x_test)
# Print results
print(confusion_matrix(y_test, y_predict))
print(classification_report(y_test, y_predict))
```
